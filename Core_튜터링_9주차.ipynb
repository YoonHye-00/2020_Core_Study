```
# 연습문제 9.8

# 클래스 정의
class Counter :
    # 초기화 메소드, 생성자
    def __init__(self, number) :
        # 이 클래스는 number라는 측정 값을 가짐 >> self.__number
        # __init__() 안에서 생성된 변수 : 인스턴스 변수
        if number >= 100 or number <= -1 :
            self.__number = 0
        else :
            self.__number = number
        
    # __number을 0으로 초기화하는 resert() 메소드를 가짐
    def resert(self) :
        self.__number = 0
    
    # __number을 1 증가시키는 inc() 메소드를 가짐
    def inc(self) :
        self.__number += 1
        if self.__number >= 100 :
            self.__number = 0
    # __number을 1 감소시키는 ded() 메소드를 가짐
    def dec(self) :
        self.__number -= 1
        if self.__number <= -1 :
            self.__number = 0
    
    # c(n)과 같이 n을 출력하는 특수 메소드를 가짐
    def __str__(self) :
        # 출력하고자 하는 값을 return 한다.
        return "c({})".format(self.__number)

# 인스턴스 c1, c2 생성
c1 = Counter(10)
c1.inc()
print('c1 =', c1)

c2 = Counter(3)
c2.inc()
c2.inc()
c2.dec()
print('c2 =', c2)
#c2.reset()
#print('c2 =', c2)
```

```
# 연습문제 9.13

# 클래스 구현
class Rectangle :
    # 클래스 변수 : 없음

    # 생성자(초기화 메소드)
    def __init__(self, x, y, width, height) :
        # 인스턴스 변수 : 생성자 안에 정의된 변수
        self.__x = x
        self.__y = y
        self.__width = width
        self.__height = height
    
    # 사각형의 좌표와 너비, 높이, 면적을 문자열로 변환
    def __str__(self) :
        return "Rectangle : (x = {0}, y = {1}, w = {2}, h = {3}, 면적 : {4})" \
        .format(self.__x, self.__y, self.__width, self.__height, self.__width * self.__height)
    
    # get, set 함수
    # 사각형의 x좌표 값을 "지정"함
    def set_x(self, x) :
        self.__x = x
    # 사각형의 x좌표 값을 "반환"함
    def get_x(self, x) :
        return self.__x

    # 사각형의 y좌표 값을 지정함
    def set_y(self, y) :
        self.__y = y
    # 사각형의 y좌표 값을 반환함
    def get_y(self,y) :
        return self.__y
    
    # 사각형의 width 값을 지정함
    def set_width(self) :
        self.__width = width  
    # 사각형의 width 값을 반환함
    def get_width(self) :
        return self.__width
    
    # 사각형의 height 값을 지정함
    def set_height(self, height) :
        self.__height = height
    # 사각형의 height 값을 반환함
    def get_height(self) :
        return self.__height

    # 현재 사각형(self)의 면적을 반환
    def area(self) :
        # 사각형의 면적 = 너비(width) * 높이(height)
        return self.__width * self.__height

    # 매개변수 r 사각형이 현재 사각형(self)과 겹치는 부분이 있으면 True 반환 
    # 겹치는 부분이 없으면 False 반환
    def overlaps(self,  r) :
        # 겹치는 것을 어떻게 아나요?
        # 우선 self 사각형의 네 좌표랑 r 사각형의 네 좌표를 구해보자!
    
        s1 = (self.__x, self.__y)
        s2 = (self.__x, self.__y - self.__height)  # y값은 - 방향으로 이동
        s3 = (self.__x + self.__width, self.__y - self.__height)
        s4 = (self.__x + self.__width, self.__y)
        s_point = [s1, s2, s3, s4]

        # r 사각형의 네 좌표
        r1 = (r.__x, r.__y)
        r2 = (r.__x, r.__y - r.__height)
        r3 = (r.__x + r.__width, r.__y - r.__height)
        r4 = (r.__x + r.__width, r.__y)
        r_point = [r1, r2, r3, r4]

        # r 사각형의 네 점 중 하나가 self 사각형 안에 들어갈까?
        for i in r_point :
            # sef 사각형의 x좌표 <= r 사각형의 한 점의 x좌표 <= self 사각형의 x좌표 + self 사각형의 너비
            if r1[0] >= self.__x and i[0] < self.__x + self.__width :
                # self 사각형의 y좌표 - self 사각형의 높이 
                # <= r 사각형의 한 점의 y좌표 <= self 사각형의 y좌표
                if (r1[1] <= self.__y) and (i[1] >= (self.__y - self.__height)) :
                    return True
        
        # self 사각형의 네 점 중 하나가 r 사각형 안에 들어갈까?
        for i in s_point :
            if i[0] >= r.__x and i[0] <= r.__x + r.__width :
                if i[1] <= r.__y and i[1] >= r.__y - r.__height :
                    return True
        
        return False

    # 매개변수 r 사각형이 현재 사각형에 포함되어 있으면 True, 아니면 False
    def contains(self, r) :
        # self, r, 사각형의 네 좌표를 찾아봅시다.
        # 우선 self 사각형의 네 좌표
        s1 = (self.__x, self.__y)
        s2 = (self.__x, self.__y - self.__height)  # y값은 - 방향으로 이동
        s3 = (self.__x + self.__width, self.__y - self.__height)
        s4 = (self.__x + self.__width, self.__y)
        s_point = [s1, s2, s3, s4]

        # r 사각형의 네 좌표
        r1 = (r.__x, r.__y)
        r2 = (r.__x, r.__y - r.__height)
        r3 = (r.__x + r.__width, r.__y - r.__height)
        r4 = (r.__x + r.__width, r.__y)
        r_point = [r1, r2, r3, r4]

        sw = 0
        # r 사각형의 네 점 중 하나가 self 사각형 안에 들어갈까?
        for i in r_point :
            # sef 사각형의 x좌표 <= r 사각형의 한 점의 x좌표 <= self 사각형의 x좌표 + self 사각형의 너비
            if r1[0] >= self.__x and i[0] < self.__x + self.__width :
                # self 사각형의 y좌표 - self 사각형의 높이 
                # <= r 사각형의 한 점의 y좌표 <= self 사각형의 y좌표
                if (r1[1] <= self.__y) and (i[1] >= (self.__y - self.__height)) :
                    sw += 1
        if sw == 4 :
            return True
        else :
            return False
```
